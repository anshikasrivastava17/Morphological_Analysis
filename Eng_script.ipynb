```python
import spacy
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tabulate import tabulate

nlp_en = spacy.load("en_core_web_sm")

sentence = "Habits often appear to make no difference until you cross a critical threshold and unlock a new level of performance."
doc_en = nlp_en(sentence)

# Extracting the words, lemmas, and POS
results_en = [[token.text, token.lemma_, token.pos_] for token in doc_en]

df_en = pd.DataFrame(results_en, columns=["Word", "Lemma", "POS"])

print(tabulate(df_en, headers="keys", tablefmt="fancy_grid"))

plt.figure(figsize=(10, 5))
sns.countplot(data=df_en, x="POS", hue="POS", palette="viridis", legend=False)
plt.title("Part of Speech Distribution - English")
plt.xticks(rotation=45)
plt.xlabel("Part of Speech (POS)")
plt.ylabel("Count")
plt.show()
```
