```python
!pip install stanza
import stanza
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate

stanza.download("hi")
nlp_hi = stanza.Pipeline("hi")

sentence_hi = "विज्ञान सत्य नहीं है। विज्ञान सत्य को खोज रहा है। जब विज्ञान ने अपनी राय बदली, उसने आपसे झूठ नहीं बोला। उसने अधिक सीखा।"
doc_hi = nlp_hi(sentence_hi)

results_hi = [[word.text, word.lemma, word.upos] for sent in doc_hi.sentences for word in sent.words]

df_hi = pd.DataFrame(results_hi, columns=["Word", "Lemma", "POS"])

print(tabulate(df_hi, headers="keys", tablefmt="fancy_grid"))

plt.figure(figsize=(10, 5))
sns.countplot(data=df_hi, x="POS", palette="plasma")
plt.title("Part of Speech Distribution - Hindi", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.xlabel("Part of Speech (POS)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
